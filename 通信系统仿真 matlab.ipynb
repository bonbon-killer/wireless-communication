## 高斯和高斯-马尔可夫过程
* 电子器件中的热噪声能够用一个高斯过程准确地建模
* 如果高斯过程通过一个线性时不变系统，其输出也是高斯过程
* 马尔可夫过程：过去对将来没有任何影响
* `高斯-马尔可夫过程`：概率密度函数为高斯型的马尔可夫过程
* 产生马尔可夫过程最简单的方法是利用递推公式：,其中wn为一个零均值、独立、同分布的随机变量，ρ是确定Xn和Xn-1之间的相关程度的一个参数


```python
# -*- encoding: utf-8 -*-
"""
@File   : 马尔可夫过程.py
@Time   : 2021/4/29 22:11
@Author : Wang
"""
import numpy as np
import matplotlib.pyplot as plt

n=1000 #序列长度
rho=0.95 #相关系数
m=50 # 自相关长度
w=np.random.normal(0,1,n)
x=np.zeros(n)
x[0]=0
for i in range(1,n):
    x[i]=rho*x[i-1]+w[i]
plt.figure()
plt.plot(x)
plt.title('Marcov-Process')
plt.xlabel('t')
Rx=np.zeros(m+1)
for i in range(m+1):
    for j in range(n-m):
        Rx[i]+=x[j]*x[j+i]
    Rx[i]=Rx[i]/(n-i)
plt.figure()
plt.plot(range(m+1),Rx)
plt.title('auto-correlation')
plt.xlabel('/tau')
plt.ylabel('R(tau)')
plt.show()

```
